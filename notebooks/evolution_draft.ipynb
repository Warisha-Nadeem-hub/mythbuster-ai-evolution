create myth dataset

```python
import pandas as pd
import numpy as np

# 1. Define our Urban Legend & Myth Dataset
myth_data = {
    "intent": [
        "chupacabra", "chupacabra", "chupacabra",
        "bermuda_triangle", "bermuda_triangle", "bermuda_triangle",
        "bloody_mary", "bloody_mary", "bloody_mary",
        "slenderman", "slenderman", "slenderman"
    ],
    "user_phrase": [
        "Have you heard about the Chupacabra goat sucker?",
        "Is the Chupacabra real or fake?",
        "Tell me about the creature attacking livestock in Puerto Rico.",
        "Why do planes disappear in the Bermuda Triangle?",
        "Is there a magnetic anomaly in the Bermuda Devil's Triangle?",
        "What is the mystery behind the Bermuda ships vanishing?",
        "What happens if I say Bloody Mary in the mirror three times?",
        "Is the Bloody Mary mirror ghost real?",
        "Tell me the legend of the mirror witch Bloody Mary.",
        "Who is Slenderman and does he live in the woods?",
        "Is Slenderman a real creature or an internet myth?",
        "Tell me about the tall featureless man in the suit."
    ]
}

# Convert to a Pandas DataFrame for easy manipulation
df = pd.DataFrame(myth_data)

# 2. Define the scientific/historical debunking responses
responses = {
    "chupacabra": "рџ”ґ DEBUNKED: The 'Chupacabra' sightings began in Puerto Rico in 1995. DNA tests on alleged carcasses have consistently proven them to be common wild coyotes or dogs suffering from severe mange, which causes hair loss and shriveled skin.",
    "bermuda_triangle": "рџ”ґ DEBUNKED: Marine insurers (like Lloyd's of London) and the US Coast Guard confirm that the number of disappearances in the Bermuda Triangle is statistically no higher than any other heavily traveled ocean region. Most incidents were caused by rogue waves or tropical storms.",
    "bloody_mary": "рџ”ґ DEBUNKED: The illusion of seeing a face change in the mirror is a psychological phenomenon called the 'Capgras Delusion' or 'Troxler's Fading.' In low light, your brain gets bored of looking at your own face and starts misinterpreting visual data as a monster.",
    "slenderman": "рџ”ґ DEBUNKED: Slenderman is 100% artificial folklore. He was created on June 10, 2009, by user Eric Knudsen (under the pseudonym Victor Surge) on the 'Something Awful' internet forum for a Photoshop ghost-photo contest."
}

print(f"Dataset loaded successfully with {len(df)} training phrases!")
```

vectorize and train model

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline

# Create a pipeline that combines Text Vectorization (TF-IDF) + ML Classifier (SVM)
# We use English stop words (like 'is', 'the') to filter out unhelpful fluff words.
model = make_pipeline(TfidfVectorizer(stop_words='english'), LinearSVC(random_state=42))

# Train the traditional ML model
model.fit(df['user_phrase'], df['intent'])
print("Traditional ML Intent Classifier trained successfully!")
```

build chatbot loop

```python
def myth_buster_bot():
    print("рџ”® Welcome to the Myth-Buster ML Bot! Type 'exit' to quit. рџ”®\n")
    print("Try asking about: The Chupacabra, Bermuda Triangle, Bloody Mary, or Slenderman.\n")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Bot: Goodbye! Keep questioning the unexplained.")
            break

        if user_input.strip() == "":
            continue

        # Predict the intent using our trained ML pipeline
        predicted_intent = model.predict([user_input])[0]

        # Retrieve the matching debunking response
        bot_response = responses[predicted_intent]

        print(f"Predicted Intent Tag: [{predicted_intent}]")
        print(f"Bot: {bot_response}\n")

# Run the chatbot
myth_buster_bot()
```

setup grok api


```python
!pip install groq getpass4 -q

import getpass
import os
from groq import Groq

# Safely prompt for your API key
if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API Key: ")

# Initialize the free LLM client
client = Groq(api_key=os.environ["GROQ_API_KEY"])
print("LLM Client initialized successfully!")
```

```python
def llm_myth_buster():
    print("рџ¤– Welcome to the LLM Generative Myth-Buster Bot! Type 'exit' to quit. рџ¤–\n")
    print("You can now ask about ANY myth, urban legend, or conspiracy theory in the world!\n")

    # This instructs the LLM on its persona and rules
    system_prompt = (
        "You are a witty, analytical, and highly objective Myth-Buster AI. "
        "Your job is to evaluate urban legends, folklore, and internet myths. "
        "Always distinguish between verified facts and fiction. Be engaging, slightly skeptical, "
        "and provide scientific or historical context where possible."
    )

    # Initialize chat history to maintain conversational memory
    conversation_history = [
        {"role": "system", "content": system_prompt}
    ]

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Bot: Goodbye! Stay skeptical.")
            break

        if user_input.strip() == "":
            continue

        # Append the user's message to the history
        conversation_history.append({"role": "user", "content": user_input})

        # Call the updated free Llama 3.1 model
        try:
            completion = client.chat.completions.create(
                model="llama-3.1-8b-instant",  # Updated supported model
                messages=conversation_history,
                temperature=0.7,               # Balanced creativity and factuality
                max_tokens=512
            )

            bot_response = completion.choices[0].message.content
            print(f"\nBot: {bot_response}\n")

            # Append the bot's response to history so it remembers next time
            conversation_history.append({"role": "assistant", "content": bot_response})

        except Exception as e:
            print(f"\nAn error occurred: {e}. Please check your API key.")

# Run the generative LLM chatbot
llm_myth_buster()
```

create private kowledge base/vectore DB

```python
from langchain_community.vectorstores import FAISS
from langchain_core.embeddings import Embeddings
from sentence_transformers import SentenceTransformer

# 1. Our specific knowledge base
knowledge_base = [
    "Myth: Polybius Arcade Game. Fact: An urban legend claims a 1981 arcade game caused amnesia and insomnia in players before vanishing. In reality, no cabinets or ROMs exist, and it was tracked back to a hoax origin on the website Coinop.org in the early 2000s.",
    "Myth: The Slender Man Stabbing. Fact: While Slenderman is fictional, a real-world crime occurred in Waukesha, Wisconsin in 2014 where two 12-year-old girls attacked a classmate, claiming they did it to appease the fictional character.",
    "Myth: The Black Eyed Children. Fact: Urban legends describe paranormal pale children with pitch-black eyes asking to enter cars or homes. It originated entirely from reporter Brian Bethel in 1996 via a ghost-story mailing list; there is zero physical evidence.",
    "Myth: The Flatwoods Monster. Fact: In 1952 in West Virginia, youths saw a glowing object crash and a robed figure with a spade-shaped head. The US Air Force investigated it as 'Project Blue Book' and concluded the crash was a meteor and the monster was a startled barn owl perched high up."
]

# 2. Correctly implement the LangChain Embeddings interface
class LangChainEmbeddingsWrapper(Embeddings):
    def __init__(self):
        self.model = SentenceTransformer('all-MiniLM-L6-v2')

    def embed_documents(self, texts):
        return self.model.encode(texts).tolist()

    def embed_query(self, text):
        # LangChain expects a single list of floats for a query
        return self.model.encode(text).tolist()

# Instantiate our compliant wrapper
embedding_wrapper = LangChainEmbeddingsWrapper()

# 3. Build the Vector Database safely
vector_db = FAISS.from_texts(texts=knowledge_base, embedding=embedding_wrapper)

print("вњ… Vector Database initialized flawlessly!")
```

RAG chatbot

```python
def rag_myth_buster():
    print("рџљЂ Welcome to the Final Phase: RAG Myth-Buster Bot! Type 'exit' to quit. рџљЂ\n")
    print("Try asking about specific niche myths like 'Polybius', 'Black Eyed Children', or 'Flatwoods'!\n")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("Bot: Goodbye! Your RAG system is fully operational.")
            break

        if user_input.strip() == "":
            continue

        try:
            # 1. RETRIEVAL
            search_results = vector_db.similarity_search(user_input, k=1)
            retrieved_context = search_results[0].page_content if search_results else "No direct database records found."

            # 2. AUGMENTATION
            rag_prompt = (
                f"You are a strict, truth-driven Myth-Buster AI.\n"
                f"Use the following Context to answer the user's question. If the answer is not in the context, "
                f"use your general knowledge but mention that it isn't verified in our secure archive.\n\n"
                f"Context:\n{retrieved_context}\n\n"
                f"User Question: {user_input}\n"
                f"Answer:"
            )

            # 3. GENERATION
            completion = client.chat.completions.create(
                model="llama-3.1-8b-instant",
                messages=[{"role": "user", "content": rag_prompt}],
                temperature=0.2,
                max_tokens=300
            )

            print(f"\nрџ”Ќ [Retrieved Context Used]: {retrieved_context}\n")
            print(f"Bot: {completion.choices[0].message.content}\n")

        except Exception as e:
            print(f"An error occurred: {e}")

# Launch the final RAG bot
rag_myth_buster()
```
